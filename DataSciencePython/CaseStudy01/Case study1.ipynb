```python
avg_income =dfc.groupby("educational_status_code")['annual_income'].mean().reset_index()
```

```python
avg_income
```

```python
def replace_income(annual_income,educational_status_code):
    if pd.isna(annual_income):
        return float(avg_income[avg_income['educational_status_code']==educational_status_code]['annual_income'])
    return annual_income
```

```python
dfc['annual_income'] =dfc.apply(lambda x: replace_income(x['annual_income']
                                                         , x['educational_status_code'])
                                                         ,axis = 1)
```

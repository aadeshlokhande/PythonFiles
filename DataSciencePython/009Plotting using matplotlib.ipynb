```python
segment = ['A','B','C','D','E','F']
revenue = np.random.randint(5000, size = 6)

exp = [0, 0, 0.15, 0, 0.1, 0]

plt.figure(figsize=(8,8))
plt.pie(revenue, 
        labels = segment,
        explode = exp,
        colors = ['r','g','b','y','grey','#ffaa2f'],
        autopct = '%1.3f%%',
        pctdistance = 1.1,
        labeldistance = 0.6,
        shadow = True)
plt.legend()
plt.show()
```

```python
#-------Getting Axes Objects for each Subplot and assigning each to different variable for subsequent usage---------------

fig, ax = plt.subplots(2, 2, sharex=True, sharey = True,figsize=(15,10))       # Assign figure object to fig and Axes objects to ax
tl = ax[0,0]                                   # Axes object for subplot 1 
tr = ax[0,1]                                   # Axes object for subplot 2
bl = ax[1,0]                                   # Axes object for subplot 3 
br = ax[1,1]                                    # Axes object for subplot 4 

#--------------------------Plotting data in Plot 1 and modifying its labels---------------------------
tl.plot(np.arange(0,40), north, label='north')                 
tl.set_title('North Sales')
tl.set_xlabel('Day_N')
tl.set_ylabel('Sale_N')

#--------------------------Plotting data in Plot 2 and modifying its labels---------------------------
tr.plot(np.arange(0,40), south, label='south', c='red')
tr.set_title('South Sales')
tr.set_xlabel('Day_S')
tr.set_ylabel('Sales_S')
    
#--------------------------Plotting data in Plot 3 and modifying its labels---------------------------
bl.plot(np.arange(0,40), east, label='east', c='green')
bl.set_title('East Sales')
bl.set_xlabel('Day_E')
bl.set_ylabel('Sales_E')
               
#--------------------------Plotting data in Plot 4 and modifying its labels---------------------------

br.set_xlabel('Day_W')
br.set_ylabel('Sales_W')
br.set_title('West Sales')
br.scatter(np.arange(0,40), north, label='north',marker='v')

#-------------------------------------------------Solution --------------------------------------------------------------
plt.tight_layout()
```
